## Methods for Creating Plot

## 1 - Functional Method

```python
plt.plot(x, y)
# plt.show()
```

##### We can also add colour to our plot

```python
plt.plot(x, y, 'r-')

# plt.plot(x, y, 'y-'), plt.plot(x, y, 'g-'), etc.
```

##### We can also label our plot

```python
plt.plot(x, y, 'g')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

#### Define subplot

```python
plt.subplot(1, 2, 1)                        # (row, column, plot_number)
plt.plot(x, y, 'r--')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b*-')
```

## 2 - Object Oriented Method

```python
fig = plt.figure()                          # Create Figure object

# Add set of axes to figure
# left, bottom, width, height (range 0 to 1)

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
```

```python
fig = plt.figure()                          # Create Figure object

# Add set of axes to figure
# left, bottom, width, height (range 0 to 1)

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Plot on that set of axes
axes.plot(x, y, 'g')
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.2, 0.4, 0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r')
axes1.set_title("Larger Plot")

axes2.plot(y, x, 'y--')
axes2.set_title("Smaller Plot")
```

### Creating Subplot

```python
fig, axes = plt.subplots(nrows=1, ncols=2)                    # No. of plots = nrows * ncols
```

```python
axes                                              # Axes is an array of axes to plot on
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, 'b')
axes[0].set_title("First Plot")

axes[1].plot(y, x, 'y')
axes[1].set_title("Second Plot")

plt.tight_layout()                       # Automatically adjust subplot parameters
```

```python
fig                                                       # Display the figure object 
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, x, 'b')
axes[0][0].set_title("First Plot")

axes[1][0].plot(x, y, 'y')
axes[1][0].set_title("Second Plot")

axes[0][1].plot(y, x**3, 'b')
axes[0][1].set_title("Fourth Plot")

axes[1][1].plot(y, x, 'y')
axes[1][1].set_title("Fifth Plot")

plt.tight_layout()          
```

# Figure Size and DPI

# Legends, labels and titles

```python
fig = plt.figure(figsize=(8, 4))

axes = fig.add_axes([0, 0, 1, 1])

# Add Two Plots
axes.plot(x, x**2, label = "X Squared")
axes.plot(x, x**3, label = "X Cubed")

axes.legend(loc = 0)   # When we call it, it looks for each plot calls and look for "label"
                       # "loc" provide the location for the labels
                       # currently labels are at top-left corner of the plot
```

# Plot Appearance

## Color

```python
fig = plt.figure(figsize=(8, 4))

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x+1, color = "green", label = "Green Line")
ax.plot(x, x+2, color = "#FF8C00", label = "Orange Line")                   # OR

ax.legend()
```

## Line Width

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, color = "#228B22", linewidth = 10)             # By default it is "1"
ax.plot(x, x**3, color = "purple", linewidth = 0.5)          # Use "lw" or "linewidth"
```

## Transparency

```python
fig = plt.figure()

axes = fig.add_axes([0,0,2,1])

axes.plot(x, x+1, linewidth=3, alpha=0.5)
axes.plot(x, x+2, linewidth=3)
```

## Linestyle

```python
fig = plt.figure()

axes = fig.add_axes([0,0,2,1])

axes.plot(x, x+1, linewidth=3, linestyle=":")
axes.plot(x, x+2, lw=3, ls="--")                               # "linestyle" == "ls"
```

## Markers

```python
# Mark the points of a graph
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,2,1])

axes.plot(x, x+1, lw=3, ls="-", marker="o")
axes.plot(x, x+2, lw=1, ls="-", marker="+", markersize=20)           # markersize

axes.plot(x, x+3, lw=1, ls="-", marker="*", 
          markersize = 20, 
          markerfacecolor = "yellow",
          markeredgewidth = 2,
          markeredgecolor = "red")
```

# Control over axis appearance

```python
fig = plt.figure()

axes = fig.add_axes([0,0,2,1])

axes.plot(x, y, color="purple", lw=2, ls="--")

axes.set_xlim([0, 1])                                       # Set Limit
axes.set_ylim([0, 2])
```
