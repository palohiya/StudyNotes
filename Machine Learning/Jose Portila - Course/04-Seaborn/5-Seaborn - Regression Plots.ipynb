# Regression Plots

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## lmplots

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')               # add 'hue'
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'])         # add 'marker'
```

```python
#Change the size of a markers
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'], scatter_kws={'s':100})
```

#### we can seperate by columns instead of colors

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

#### aspect and size ratio

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex',
           aspect=0.6, size=8)
```
