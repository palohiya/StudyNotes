# Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)                                  # Pair Plot
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

## PairGrid

#### Unlike pairplot, in PairGrid we have full control on our plots

We get empty plots where we can draw according to our choices.

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

##### Combining many different plots

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)                                  # Diagonal Plots
g.map_upper(plt.scatter, color='red')                    # Upper Diagonal Plots
g.map_lower(sns.kdeplot, color='green')                 # Lower Diagonal Plots
```

# Facet Grid

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```
