# Seaborn

Seaborn is a Python data visualization library based on matplotlib. 

It provides a high-level interface for drawing attractive and informative statistical graphics. 

Seaborn helps resolve the two major problems faced by Matplotlib; the problems are ?

    Default Matplotlib parameters
    Working with data frames

```python
import seaborn as sns
```

```python
%matplotlib inline
```

### Create DataFrame

```python
tips = sns.load_dataset('tips')             # 'tips' is a built-in data set
```

```python
tips.head()
```

```python
type(tips)
```

# Plot Types

## distplot

The distplot shows the distribution of a univariate set of observations.

It provide us a histogram.

```python
sns.distplot(tips['total_bill'])       # we pass single column as a dataset
```

```python
# kernel density estimate non-parametric way to estimate the probability density function of a random variable
```

```python
sns.distplot(tips['total_bill'], kde=False)            # kde = kernel density estimate
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)          # bins = Specification of hist bins
```

## jointplot

It draw a plot of two variables or columns.

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

All kinds : { "scatter" | "reg" | "resid" | "kde" | "hex" }

```python
# By default it is "scatter"
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex') # We can pass 'hex' in it to get hexagon representation
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='resid')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

## pairplot

Create plots for all the numeric columns values.

```python
sns.pairplot(tips)
```

```python
# We see histogram instead of 'scatter' plot when we compare the column to itself.
```

```python
sns.pairplot(tips, hue='sex') # we can use "hue" for columns for non-numeric values
                              # it plot the graphs with those non-numeric values
```

## rugplot

For a single quantitative variable.

```python
#Shows point where the valuse really is.
```

```python
sns.rugplot(tips['total_bill'])
```

## kdeplot

##### Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable.

These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value.

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```

```python
sns.kdeplot(tips['tip'])
```

# Categorical Plots

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## barplot

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

## countplot

```python
# Count the occurance
```

```python
sns.countplot(x='sex', data=tips)
```

## boxplot

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

## violinplot

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

## stripplot

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

## swarmplot

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

# Combining Categorical Plots

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color='white')
```

## factorplot

```python
sns.factorplot(x='sex', y='total_bill', data=tips, kind='bar')
```

```python
sns.factorplot(x='sex', y='total_bill', data=tips, kind='violin')
```

```python
sns.factorplot(x='sex', y='total_bill', data=tips, kind='swarm')
```
